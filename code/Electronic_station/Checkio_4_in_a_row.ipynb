```python
def checkio(matrix):
    count = [1] * (2*len(matrix)-1)
    k = count[:]
    N = len(matrix)
    for i, row in enumerate(matrix):   #horizontal
        for j, elem in enumerate(row[:-1]):
            if matrix[i][j] == matrix[i][j+1]:
                count[i] += 1
                if count[i] >= 4:
                    return True
            else: 
                count[i] = 1
    count = k[:]
    
    for i, row in enumerate(matrix[:-1]):      #vertical
        for j, elem in enumerate(row):
            if matrix[i][j] == matrix [i+1][j]:
                count[j] += 1
                if count[j] >= 4:
                    return True
            else: 
                count[j] = 1
    count = k[:]
          
    for l in range(2):                         #diag \\\ and then \\\
        for i, row in enumerate(matrix[:-1]):  
            for j, elem in enumerate(row):
                if (i + j + 1 >= 4) and (i + j + 1 <= N*2-4):
                    if i - 1 >= 0 or j - 1 >= 0:
                        if matrix[i][j] == matrix[i+1][j-1]:
                            count[i+j] += 1
                            if count[i+j] >= 4:
                                return True
                        else:
                            count[i+j] = 1
        count = k[:]
        for row in matrix:
            row = row.reverse()
    
    return False

if __name__ == '__main__':
    #These "asserts" using only for self-checking and not necessary for auto-testing
    
    assert checkio([
        [7, 1, 1, 8, 1, 1],
        [1, 1, 7, 3, 1, 5],
        [2, 3, 1, 2, 5, 1],
        [1, 1, 1, 5, 1, 4],
        [4, 6, 5, 1, 3, 1],
        [1, 1, 9, 1, 2, 1]
    ]) == True, "Diagonal"
    
    assert checkio([
        [1, 2, 1, 1],
        [1, 1, 4, 1],
        [1, 3, 1, 6],
        [1, 7, 2, 5]
    ]) == True, "Vertical"
    
    assert checkio([
        [7, 1, 4, 1],
        [1, 2, 5, 2],
        [3, 4, 1, 3],
        [1, 1, 8, 1]
    ]) == False, "Nothing here"
    
    assert checkio([
        [2, 1, 1, 6, 1],
        [1, 3, 2, 1, 1],
        [4, 1, 1, 3, 1],
        [5, 5, 5, 5, 5],
        [1, 1, 3, 1, 1]
    ]) == True, "Long Horizontal"

    assert checkio([
        [2, 6, 2, 2, 7, 6, 5],
        [3, 4, 8, 7, 7, 3, 6],
        [6, 7, 3, 1, 2, 4, 1],
        [2, 5, 7, 6, 3, 2, 2],
        [3, 4, 3, 2, 7, 5, 6],
        [8, 4, 6, 5, 2, 9, 7],
        [5, 8, 3, 1, 3, 7, 8]
    ]) == False, "Diagonal"
```
